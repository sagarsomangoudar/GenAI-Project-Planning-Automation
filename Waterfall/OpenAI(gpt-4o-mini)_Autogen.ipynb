```python
import os
import openai
import autogen
from pydantic import BaseModel
from constants import Constants

os.environ["AUTOGEN_USE_DOCKER"] = "0"

openai.api_key = Constants.OPENAI_API_KEY

client = openai.OpenAI(api_key = Constants.OPENAI_API_KEY)
llm_config={"model": "gpt-4o-mini",  
            "api_key": Constants.OPENAI_API_KEY} 

task = "I want to develop, test and deploy a web and Artificial Intelligence based application foor our retail Glocery store where my customers can browse, search an item, scan any item using barcode, add to cart, online payment, online-delivery, tracking of order and online feedback. You have to do the project in a Waterfall Model. " + "You have to give me the estimates from every phase. seperately"
```

```python
config_list = [
  {
    "model": "gpt-4o-mini",
    "api_key": "API_KEY",       # Put OpenAI API key here
    "tags": [
      "gpt-4o-mini",
      "tool",
      "vision"
    ]
  }
]

customerGuy = autogen.ConversableAgent(
    name="Customer",
    llm_config={"config_list": config_list},
    system_message="""
    You are retail store manager, You know every corner of your retail store.
    You can transform your complex concepts as requirement to a IT company Requirement Enginner.
    You should improve the quality of the content based on the feedback from the user.
    """,
    code_execution_config = False, 
    human_input_mode = "NONE",
)

projectManager = autogen.ConversableAgent(
    name="Project-Manager",
    llm_config={"config_list": config_list},
    system_message="""
   You are a project manager of this project. Your role is to oversee project planning, execution, and completion. They define project scope, create timelines, allocate resources, and manage budgets. They coordinate teams, communicate with stakeholders, and ensure projects meet quality standards and deadlines while mitigating risks and resolving issues efficiently.
   You will take the problem statement from the customer agent and start planning with requirement engineers
    """,
    code_execution_config = False, 
    human_input_mode = "NONE",
)
```

```python
initiatedChat = customerGuy.initiate_chat(
    projectManager,
    message=task,
    max_turns=2, type = 'model_error', param = None, code =  None,
        speaker_selection_method = 'round_robin' , allow_repeat_speaker = False, 
)
```
